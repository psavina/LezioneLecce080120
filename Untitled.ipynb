## Photon - Hadron primary Cosmic Ray discrimination using ML

### 1. Data

The signatures of a photon induced air shower are a large atmospheric depth at the shower maximum ($X_{\text{max}}$) and a lower number of muons that reach the ground ($N_{\mu}$) with respect to the bulk of hadron-induced background. 

<img src="images/diffPrim.png" width="60%"/>

Root files with signals (photon) and background (proton) simulations can be found here:

- [Protons](http://www.dmf.unisalento.it/~lorenzo/LabAnalisi2019/photon_hadron/conex_eposlhc_1301560476_100.root)
- [Photons](http://www.dmf.unisalento.it/~lorenzo/LabAnalisi2019/photon_hadron/conex_eposlhc_2086832343_0.root)

Download the files and put them all under a files/ directory.

Each root file contains 2 TTrees with name "Header" and "Shower". 

We will use the TTree named **Shower** that is filled with a set of variables the describes the shower development.



### 2. Load Data

```python
import uproot
import numpy as np
import pandas as pd
```

```python
filesDict = {
    "protons": "files/conex_eposlhc_1301560476_100.root",
    "photons": "files/conex_eposlhc_2086832343_0.root"
}

dfDict = {} # dataframe dict

for key in filesDict: #for each primary type
    print('loading'+filesDict[key]+'...')
    
    #flatten=False because our tree contains arrays
    dfDict[key] = uproot.open( filesDict[key] )['Shower'].pandas.df(flatten=False)
    
    #add a column with an int that declares whether the event is signal(=1) or backround (=0).
    if key == 'photons':
        dfDict[key]['isSignal'] = 1
    else:
        dfDict[key]['isSignal'] = 0
        
    # log of the number of muons in the final step  
    dfDict[key]['lgM'] = np.log(dfDict[key]['Mu'].str[-1])

dfDict['photons'].head()
#dfDict['protons'].head()
```

### 3. Data Set

In the **Shower** TTree several parameters can be identified.
Some of them are properties of the primary particle, the others
describe the shower development as a function of the grammage $X$.

The most usefull in our case are:

- lgE: $\log_{10} \left( E/\mathrm{[eV]} \right)$;
- nX: number of grammage levels considered in the calculations;
- Xmx: $X_{\text{max}}$ $[\mathrm{gr} / \mathrm{cm}^2]$;
- Mu[i]: number of muons at the $i$-th grammage value.

#### Preparing data as inputs to a DNN

```python
for key in dfDict:
    s  = "files/"
    s += key
    s += ".pkl"
    dfDict[key].to_pickle(s)
```

#### Preparing signal and background inputs and data shuffling

```python
import sklearn.utils

def Shuffling(df, mSeed=130392):
    df = sklearn.utils.shuffle(df,random_state=mSeed)
    df = df.reset_index(drop=True)
    return df

sDfDict = {}

for key in dfDict:
    sDfDict[key] = Shuffling(dfDict[key])
    
    
signList = [sDfDict['photons'][:5000]]
dfSignal = pd.concat( signList, ignore_index=True)
backList = [sDfDict['protons'][:5000]]
dfBackgr = pd.concat( backList, ignore_index=True)

dfTotal = Shuffling( pd.concat( [dfSignal, dfBackgr], ignore_index=True) )
print ("Total events after shuffling: ",dfTotal.shape)
dfTotal.to_pickle( 'files/mixData.pkl' )
```

### 4. Building and Training a Deep Neural Network with Keras

```python
inputs = ['Xmx','lgM']

X = dfTotal[inputs].values
tmpY = dfTotal['isSignal']
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
scaler = StandardScaler()
X = scaler.fit_transform(X)

le = LabelEncoder()
y = le.fit_transform(tmpY)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7,random_state=123)

X_train.shape
X_test.shape
```

```python
from keras.models import Model, Sequential
from keras.layers import Dense, Dropout, Input, BatchNormalization
from keras.callbacks import EarlyStopping, ModelCheckpoint
from keras.layers.core import Dense, Activation

def BuildDNN(N_input,width,depth):
    model = Sequential()
    model.add(Dense(units=width, input_dim=N_input))  # First layer
    model.add(Activation('relu'))
    model.add(Dropout(0.2))

    for i in range(0, depth):
        model.add(Dense(width))
        model.add(Activation('relu'))
        # Dropout randomly sets a fraction of input units to 0 at each update during training time
        # which helps prevent overfitting.
        model.add(Dropout(0.2)) 

    model.add(Dense(1, activation='sigmoid'))  # Output layer/node 

    return model


n_dim=X_train.shape[1]
n_nodes = 32
n_depth = 2 # number of additional hidden layers
```

```python
model=BuildDNN(n_dim,n_nodes,n_depth)
```

```python
model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
```

```python
modelMetricsHistory = model.fit(X_train, y_train, epochs=20,batch_size=2048,validation_split=0.2, verbose=1)
```

```python
perf = model.evaluate(X_test, y_test, batch_size=2048)
testLoss = 'Test loss:%0.3f' % perf[0]
testAccuracy = 'Test accuracy:%0.3f' % perf[1]
print (testLoss)
print (testAccuracy)
```

```python
# summarize history for accuracy
plt.plot(modelMetricsHistory.history['acc'])
plt.plot(modelMetricsHistory.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.figtext(0.5, 0.5, testAccuracy, wrap=True, horizontalalignment='center', fontsize=10)
plt.savefig("files/Accuracy.png")
plt.show()
plt.clf()
# summarize history for loss
plt.plot(modelMetricsHistory.history['loss'])
plt.plot(modelMetricsHistory.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.figtext(0.5, 0.5, testLoss, wrap=True, horizontalalignment='center', fontsize=10)
plt.savefig("files/Loss.png")
plt.show()
plt.clf()
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score, classification_report

# Generates output predictions for the input samples.
yhat_test = model.predict(X_test,batch_size=2048)

print (yhat_test)
```

```python
# Get 'Receiver operating characteristic' (ROC)
fpr, tpr, thresholds = roc_curve(y_test, yhat_test)

# Compute Area Under the Curve (AUC) from prediction scores
roc_auc  = auc(fpr, tpr)
print ("ROC AUC: %0.3f" % roc_auc)

plt.plot(fpr, tpr, color='darkorange',  lw=2, label='Full curve (area = %0.2f)' % roc_auc)
plt.plot([0, 0], [1, 1], color='navy', lw=2, linestyle='--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.title('ROC curves for Signal vs Background')
# plt.plot([0.038], [0.45], marker='*', color='red',markersize=5, label="Cut-based",linestyle="None")
# plt.plot([0.038, 0.038], [0,1], color='red', lw=1, linestyle='--') # same background rejection point
plt.legend(loc="lower right")
# plt.savefig("Files/ROC.pdf")
plt.show()
plt.clf()
```

```python
from keras import constraints

def BuildDNN_dropout(N_input,width,depth):
    model = Sequential()
    model.add(Dense(units=width, input_dim=N_input, kernel_constraint=constraints.max_norm(3) ) )  # First layer
    model.add(Activation('relu'))
    model.add(Dropout(0.2))

    for i in range(0, depth):
        model.add(Dense(width))
        model.add(Activation('relu'))
        # Dropout randomly sets a fraction of input units to 0 at each update during training time
        # which helps prevent overfitting.
        model.add(Dropout(0.2)) 

    model.add(Dense(1, activation='sigmoid'))  # Output layer/node 

    return model
```

```python
model_dro=BuildDNN_dropout(n_dim,n_nodes,n_depth)
model_dro.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['accuracy'])
# Start the training!
history_dro = model_dro.fit(X_train, y_train, epochs=20,batch_size=2048,validation_split=0.2, verbose=1)
```

```python
# summarize history for accuracy
plt.plot(history_dro.history['acc'])
plt.plot(history_dro.history['val_acc'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='lower right')
plt.figtext(0.5, 0.5, testAccuracy, wrap=True, horizontalalignment='center', fontsize=10)
plt.savefig("files/Accuracy.png")
plt.show()
plt.clf()
# summarize history for loss
plt.plot(history_dro.history['loss'])
plt.plot(history_dro.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper right')
plt.figtext(0.5, 0.5, testLoss, wrap=True, horizontalalignment='center', fontsize=10)
plt.savefig("files/Loss.png")
plt.show()
plt.clf()
```

```python
callbacks = [
# if we don't have an increase of the accuracy for 10 epochs, terminate training.
EarlyStopping(verbose=True, patience=10, monitor='val_loss'),
# Always make sure that we're saving the model weights with the best loss.
ModelCheckpoint('model.h5', monitor='val_loss', verbose=True, save_best_only=True)
]

model_dro =BuildDNN_dropout(n_dim,n_nodes,n_depth)
model_dro.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
# Start the training!
history_dro = model_dro.fit(X_train, y_train, callbacks=callbacks, epochs=200,batch_size=2048,validation_split=0.2, verbose=1)
```

### K-fold cross validation

```python
from sklearn.model_selection import StratifiedKFold
from collections  import Counter
from keras import backend as K


df_tot = dfTotal

X=df_tot[inputs]
scaler = StandardScaler()
X = scaler.fit_transform(X)

Y= df_tot['isSignal']


kfold = StratifiedKFold(n_splits=4, shuffle=False, random_state=None)
cv_scores = []
ii = 1
tprs = []
aucs = []
mean_fpr = np.linspace(0, 1.0, 100)
    
for train_index, test_index in kfold.split(X, Y):
    print('{} of KFold {}'.format(ii,kfold.n_splits))
    
    X_train_xval = np.array(X)[train_index,:]
    X_test_xval = np.array(X)[test_index,:]
    y_train_xval = np.array(Y)[train_index]
    y_test_xval = np.array(Y)[test_index]
    
    print (X_train_xval.shape[1])

    model = BuildDNN_dropout(X_train_xval.shape[1],32,3)
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    
    # adjust the learning rate
    K.set_value(model.optimizer.lr, 0.003 )
    
    callbacks = [
            EarlyStopping(verbose=True, patience=5, monitor='val_loss'),
            ModelCheckpoint('model_kfold_'+str(ii)+'.h5', monitor='val_loss', verbose=True, save_best_only=True)
    ]

    kf_history = model.fit( X_train_xval, y_train_xval, epochs=200, batch_size=2048,
            validation_split=0.2,
            callbacks=callbacks,
            verbose=1)
    
    kf_y_prediction_test = model.predict(X_test_xval)
    score = roc_auc_score(y_test_xval,kf_y_prediction_test)
    print('ROC AUC score:',score)
    cv_scores.append(score)    
    
    # Get 'Receiver operating characteristic' (ROC)
    fpr, tpr, thresholds = roc_curve(y_test_xval, kf_y_prediction_test)
    roc_auc = auc(fpr, tpr)
    tprs.append(np.interp(mean_fpr, fpr, tpr))
    tprs[-1][0] = 0.0
#     print(roc_auc)
    aucs.append(roc_auc)
    plt.plot(fpr, tpr, lw=1, alpha=0.3,label='ROC fold %d (AUC = %0.3f)' % (ii, roc_auc))

    ii += 1

plt.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',label='Luck', alpha=.8)
mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
plt.plot(mean_fpr, mean_tpr, color='b',label=r'Mean ROC (AUC = %0.3f $\pm$ %0.3f)' % (mean_auc, std_auc),lw=1, alpha=.7)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
plt.fill_between(mean_fpr, tprs_lower, tprs_upper, color='grey', alpha=.2, label=r'$\pm$ 1 std. dev.')

plt.xlim([-0.05, 1.05])
plt.ylim([-0.05, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
# plt.title('Receiver operating characteristic example')
plt.xticks(np.arange(0.0, 1.1, 0.1))
plt.yticks(np.arange(0.0, 1.1, 0.1))
plt.title('ROC curves for Signal vs Background')
plt.legend(loc="lower right",fontsize='x-small')
plt.show()
```
